```python
# Regressão Polinomial
# Importando as bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Selecionando as colunas de interesse para a análise

X = dados_gpssat3[['Apont de Operação (Hs)']]
y = dados_gpssat3['Tempo de Motor em Apont Operação (Hs)']

# Dividindo os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Criando o modelo de regressão polinomial
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X_train)

# Treinando o modelo de regressão linear com as variáveis polinomiais
regressor = LinearRegression()
regressor.fit(X_poly, y_train)

# Fazendo as predições com o conjunto de teste
X_test_poly = poly_reg.fit_transform(X_test)
y_pred = regressor.predict(X_test_poly)

# Avaliando o modelo utilizando o coeficiente de determinação (R²)
r2 = r2_score(y_test, y_pred)
print('Coeficiente de determinação (R²):', r2)

# Imprimindo os coeficientes a0, a1 e a2
a0 = regressor.intercept_
a1, a2 = regressor.coef_[1:]

print('Coeficientes do polinômio:')
print(f'a0 = {a0:.2f}, a1 = {a1:.2f}, a2 = {a2:.2f}')


# Plotando os dados de treinamento
plt.scatter(X_train, y_train, color='blue')

# Ordenando os valores de x antes de plotar a linha da regressão
sorted_x = np.sort(X_train, axis=None)

# Prevendo os valores de y para os valores ordenados de x
y_pred = regressor.predict(poly_reg.fit_transform(sorted_x.reshape(-1, 1)))

# Plotando a linha da regressão polinomial
plt.plot(sorted_x, y_pred, color='red')

# Configurando o título e os rótulos dos eixos
plt.title('Regressão polinomial')
plt.xlabel('Apont de Operação (Hs)')
plt.ylabel('Tempo de Motor em Apont Operação (Hs)')

# Exibindo o gráfico
plt.show()
```

```python
# Regressão Linear
# Importando as bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Selecionando as colunas de interesse para a análise
X = dados_gpssat3[['Apont de Operação (Hs)']]
y = dados_gpssat3['Tempo de Motor em Apont Operação (Hs)']

# Dividindo os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Treinando o modelo de regressão linear
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Fazendo as predições com o conjunto de teste
y_pred = regressor.predict(X_test)

# Coeficientes da regressão linear
a = regressor.coef_[0]
b = regressor.intercept_

print('Coeficiente angular (a):', a)
print('Coeficiente linear (b):', b)

# Avaliando o modelo utilizando o coeficiente de determinação (R²)
r2 = r2_score(y_test, y_pred)
print('Coeficiente de determinação (R²):', r2)

# Plotando os dados de treinamento
plt.scatter(X_train, y_train, color='blue')

# Plotando a linha de regressão
plt.plot(X_train, regressor.coef_*X_train + regressor.intercept_, color='red')

# Adicionando título e rótulos dos eixos
plt.title('Regressão Linear Simples')
plt.xlabel('Apont de Operação (Hs)')
plt.ylabel('Tempo de Motor em Apont Operação (Hs)')

# Mostrando o gráfico
plt.show()
```
